```python
import pandas as pd
import plotly as pt
from  geopy.geocoders import Nominatim
from tqdm import tqdm
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected = True)
```

```python
#https://taggs.hhs.gov/Coronavirus/Providers
df = pd.read_csv('HHS_Provider_Relief_Fund.csv')
df
```

```python
df['Payment'] = pd.to_numeric(df['Payment'].str[1:].str.replace(',',''))
df.dtypes
```

```python
df_state_sum = df[['State', 'Payment']].groupby('State').agg(['describe','sum'])
df_state_sum
```

```python
df_state_num_final=df_state_sum.droplevel([0,1], axis=1) 
df_state_num_final.reset_index(inplace=True)
df_state_num_final.rename(columns={'Payment': 'sum'}, inplace=True)
```

```python
df_state_num_final
```

```python
import plotly.express as px
graphs = ['sum', 'count', 'mean']
for idx, heat in enumerate(graphs):
    fig = px.choropleth(df_state_num_final,
                    locations='State',
                    color=heat,
                    color_continuous_scale='spectral_r',
                    hover_name='State',
                    locationmode='USA-states',
                    labels={'Current Unemployment Rate':'Unemployment Rate %'},
                    scope='usa',
                   hover_data=['sum', 'count', 'mean'])
    print(heat)
    fig.show()
    #iplot(fig)
```

```python
df_state_city = df[['State','City','Payment']].groupby(['State','City']).sum()
df_state_city.reset_index(inplace=True)
df_state_city
```

```python
df_state_city['location'] = df_state_city['City'] + ', ' + df_state_city['State'] + ', United States'
df_state_city
```

```python
geolocator = Nominatim(user_agent="Test")
location = geolocator.geocode('ANCHOR POINT, AK, United States')
print(location.longitude)
print(location.latitude)
```

```python
import warnings
warnings.filterwarnings(action='ignore')
tqdm.pandas()
from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
df_state_city['geo'] = df_state_city['location'].progress_apply(geocode)
df_state_city
```

```python
df_state_city_none=df_state_city.loc[df_state_city['geo'].astype(str) == 'None']
df_state_city_none
```

```python
df_state_city_complete = df_state_city.loc[df_state_city['geo'].astype(str) != 'None']
df_state_city_complete
```

```python
df_state_city_complete['lat'] = df_state_city_complete['geo'].progress_apply(lambda x: x.latitude)
df_state_city_complete['lon'] = df_state_city_complete['geo'].progress_apply(lambda x: x.longitude)
df_state_city_complete.to_csv('back_up_geo.csv')
df_state_city_complete
```

```python
df_state_city_complete = pd.read_csv('back_up_geo.csv')
df_state_city_complete
```

```python
import plotly.io as pio
pio.renderers.default = 'iframe'
```

```python
import plotly.graph_objects as go
limits = [(0,100000),(100001,2000000),(2000000,10000000),(10000001,100000000),(100000001,1000000000)]
colors = ["lightgrey","lightseagreen","yellow","blue","red"]
fig = go.Figure()
scales = [40000, 200000, 400000, 800000, 1000000]
for i, lim in enumerate(limits):
    df_sub = df_state_city_complete.loc[df_state_city_complete['Payment'].between(lim[0], lim[1])]
    fig.add_trace(go.Scattergeo(
            locationmode = 'USA-states',
            lon = df_sub['lon'],
            lat = df_sub['lat'],
            text = df_sub['location'],
            marker = dict(
                size = df_sub['Payment']/scales[i],
                line_color='rgb(40,40,40)',
                color = colors[i],
                line_width=0.5,
                sizemode = 'area'
            ), name = f'{lim[0]} - {lim[1]}'))
fig.update_layout(
        title_text = 'COVID19 PROVIDER RELIEF FUND PAYMENTS (CITIES)',
        showlegend = True,
        geo = dict(
            scope = 'usa',
            landcolor = 'rgb(217, 217, 217)',
        )
    )
fig.show()
```
